```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df=pd.read_csv(r'sample_data_100_records.csv')

#Using head,tail and info function
print(df.head())
print(df.tail())
print(df.info())

#slicing and dicing of data
print(df.loc[0:5,'Name'])

#filter data using boolean indexing
print(df[df['Name']=="Sophia White"])

#use groupby function
print(df.groupby('Name').count())
print(df.groupby('Age').sum())

#visualize the data
plt.plot(df.groupby('Age').count())
plt.show()

#split the data into test and training data
from sklearn.model_selection import train_test_split
train, test=train_test_split(df,test_size=0.2)
print(train)

#merge the data
df1=pd.read_csv(r'sample_data_100_records.csv')
df2=pd.read_csv(r'sample_data_100_records.csv')
print(pd.merge(df1,df2,on='Name'))

#dictionary of lists
dict = {
    "First Score": [100, 90, np.nan, 95],
    "Second Score": [30, 45, 56, np.nan],
    "Third Score": [np.nan, 40, 80, 98],
}

#creating dataframe from list
df = pd.DataFrame(dict)

#using isnull() function
df.isnull()
```
